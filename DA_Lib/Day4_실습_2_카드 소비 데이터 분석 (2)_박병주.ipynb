  - \t으로 분리된 csv 파일
  - 전처리된 데이터가 저장되어 있음

## 1. 데이터 로딩 to DataFrame
- 전처리하여 csv로 저장한 파일 '201906.csv'을 판다스 dataframe으로 로딩 (encoding='utf-8', index_col=0)
- 앞부분 데이터 확인
- 전체 데이터 수 확인

```python
import pandas as pd
bc_card = pd.read_csv('bc_card_out2020_03.txt', sep='\t', encoding='utf-8')
bc_card
```

```python
# 서울에 거주 중인 고객데이터 추출
index1 = bc_card[bc_card['CSTMR_MEGA_CTY_NM'] != '서울특별시'].index
seoul_card_df = bc_card.drop(index1)
seoul_card_df.reset_index(drop=True, inplace=True)
seoul_card_df
```

```python
# 서울에 비거주 중인 고객데이터 추출
index2 = bc_card[bc_card['CSTMR_MEGA_CTY_NM'] == '서울특별시'].index
# 해당 index 데이터 삭제한 df 저장
not_seoul_card_df = bc_card.drop(index2)
# 새로 생성한 df의 index 재정렬
not_seoul_card_df.reset_index(drop=True, inplace=True)
not_seoul_card_df
```

## 2. 서울시 거주/비거주 고객의 소비액 구하기
* 서울시 거주 고객 수와 소비액
    - 남성 고객수, 소비액
    - 여성 고객수, 소비액
* 서울시 비거주 고개 수와 소비액
    - 남성 고객수, 소비액
    - 여성 고객수, 소비액

```python
seoul= pd.pivot_table(seoul_card_df, values=['AMT'], aggfunc='sum', index=['SEX_CTGO_CD'])
print('서울시 거주 고객의 고객수와 소비액 현황')
print(f"남성 : {len(seoul_card_df[seoul_card_df['SEX_CTGO_CD'] == 1]):,}명, {seoul.loc[1, 'AMT']:,}원")
print(f"여성 : {len(seoul_card_df[seoul_card_df['SEX_CTGO_CD'] == 2]):,}명, {seoul.loc[2, 'AMT']:,}원")
```

```python
not_seoul = pd.pivot_table(not_seoul_card_df, values=['AMT'], aggfunc='sum', index=['SEX_CTGO_CD'])
print('비서울시 거주 고객의 고객수와 소비액 현황')
print(f"남성 : {len(not_seoul_card_df[not_seoul_card_df['SEX_CTGO_CD'] == 1]):,}명, {not_seoul.loc[1, 'AMT']:,}원")
print(f"여성 : {len(not_seoul_card_df[not_seoul_card_df['SEX_CTGO_CD'] == 2]):,}명, {not_seoul.loc[2, 'AMT']:,}원")
```

### 3. 서울시 거주자 편의점 소비 건수와 소비액 구하기

```python
seoul_cov = seoul_card_df[seoul_card_df['TP_BUZ_NO'] == 4010]
print('서울시 거주자 편의점 소비 건수와 소비액 현황')
print(f"소비 건수 : {seoul_cov['CNT'].sum(axis=0):,} , 소비액 : {seoul_cov['AMT'].sum(axis=0):,}")
```

### 4. 강남구 거주자의 강남구 내 소비액과 강남구 외 소비액 구하기

```python
gangnam_gangnam = seoul_card_df[(seoul_card_df['CSTMR_CTY_RGN_NM'] == '강남구') & (seoul_card_df['CTY_RGN_NM'] == '강남구')]
gangnam_non_gangnam = seoul_card_df[(seoul_card_df['CSTMR_CTY_RGN_NM'] == '강남구') & (seoul_card_df['CTY_RGN_NM'] != '강남구')]
print("강남구 거주자의 강남구 내 소비액과 강남구 외 소비액 현황")
print(f"강남구 거주자의 강남구 내 소비액 : {gangnam_gangnam['AMT'].sum(axis=0):,}")
print(f"강남구 거주자의 강남구 외 소비액 : {gangnam_non_gangnam['AMT'].sum(axis=0):,}")
```
