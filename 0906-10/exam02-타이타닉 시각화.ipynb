```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
row_data = pd.read_csv('data/titanic.csv', sep='\t')
row_data
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12, 6))

ax[0].set_title('Survived')
row_data['Survived'].value_counts().plot.pie(autopct='%.1f%%', ax=ax[0])
ax[0].set_ylabel('')
row_data['Survived'].value_counts().plot.bar(ax=ax[1])
plt.rc('font', family='gulim')
ax[1].set_ylabel('ëª…')
ax[1].set_title('Servived')
sns.countplot(x='Survived', data=row_data, ax=ax[2])
ax[2].set_title('Servived')
```
