```python
import numpy as np

def activaton(s):
    if s > 0: return 1
    else : return -1

def perceptron(x1, x2):
    x = np.array([x1, x2])
    tmp = np.sum(W * x) + b
    return activaton(tmp)
```

```python
print("--- 퍼셉트론으로 구현한 AND 회로 ---")
W = np.array([0.5, 0.5])
b = -0.7

for x1, x2 in [(-1, -1), (-1, 1), (1, -1), (1, 1)]:
    y = perceptron(x1, x2)
    print('{:3d} {:3d} : {:3d}'. format(x1, x2, y))
```
