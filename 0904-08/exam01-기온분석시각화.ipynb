# 매년 8월의 최고온도를 시각화

```python
with open('C:/Lecture/python/data/seoul-from(1976-2023).csv', 'r') as file:
    data = csv.reader(file)
    next(data)
    highTempList = list()
    for row in data:
        if len(row) > 1 and row[-1] != '':
            date = row[0].split('-')
            month = date[1]
            if month == '08':
#                 print(row)
                highTempList.append(float(row[-1]))
plt.figure(figsize=(10,3))
plt.plot(highTempList, 'r')
```

# 매년 생일의 최저온도와 최고 온도의 시각화

```python
with open('C:/Lecture/python/data/seoul-from(1976-2023).csv') as file:
    data = csv.reader(file)
    next(data)
    highTempList = []
    lowTempList = []
    years = []
    for row in data:
#         print(row)
        if len(row) > 1 and row[-1] !='' and row[-2] != '':
            date = row[0].split('-')
#             print(row)
            month = date[1]
            date2 = date[2]
            year = date[0]
#             print(year)
            if date2 =='27' and month == '08':
                highTempList.append(float(row[-1]))
                lowTempList.append(float(row[-2]))
                years.append(date[0].strip())
#                 print(row)
# print(highTempList)
# print(lowTempList)

plt.figure(figsize=(15,5))
plt.xticks(rotation = 90)
plt.plot(years, highTempList)
plt.plot(years, lowTempList)
```
