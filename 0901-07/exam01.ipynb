```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10,20,30,40])
```

```python
plt.plot([10,20,30,40], [12,45,23,87])
```

```python
plt.plot([10,20,30,40], label='asc', color='red', linestyle='dashed')
plt.plot([40,30,20,10], 'red', label='desc', ls='dotted')
plt.legend(loc=5)
```

```python
plt.plot([10,20,30,40], label='asc', color='r', linestyle='--')
plt.plot([40,30,20,10], 'g', label='desc', ls=':')
plt.legend(loc=5)
```

```python
plt.plot([10,20,30,40], 'r.', ls=':')
plt.plot([40,30,20,10], 'b^', ls='-.')
```
