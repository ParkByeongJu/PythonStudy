```python
from time import time
n = 1000000
x = np.random.random(n).tolist()
tic = time()
y = []
for i in range(0, n, 1):
    y.append(x[i]*x[i])
toc = time()
print("list time =", (toc-tic)*1000, "ms")
```

```python
from time import time
n = 1000000
x = np.random.random(n)
tic = time()
y = x**2
toc = time()
print("tensor time =", (toc-tic)*1000, "ms")
```
