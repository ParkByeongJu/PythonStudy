```python
import csv
import matplotlib.pyplot as plt
```

### 방법1

```python
with open('data/202308_gender.csv', 'r') as file:
    data = csv.reader(file)
    header = next(data)
    print(header[3], header[3+100], header[106], header[106+100])
    town = '성복'
    for row in data: 
        if town in row[0]:
            male = list()
            female = list()
            for i in range(0, 101):
                male.append(int(row[i+3].replace(',', '')))
                female.append(int(row[i+106].replace(',', '')))            
    print(male)
    print(female)
```

# 남성인구수를 음수로 변환

```python
for i in range(len(male)):
    male[i] = -male[i]
```

### 시각화

```python
plt.rc('font', family='NanumGothic')
plt.title(f'{town} 연령 성별 그래프')
plt.barh(range(101), male, label='male')
plt.barh(range(101), female, label='female')
plt.legend()
```

### 해당 동의 남여 비율 원그래프 시각화

```python
with open('data/202308_gender.csv', 'r') as file:
    data = csv.reader(file)
    header = next(data)
    town = '가락2동'
    for row in data: 
        if town in row[0]:
            male = [int(i.replace(',', ''))for i in row[3:104]]
            female = [int(i.replace(',', ''))for i in row[106:]]
    male = [-i for i in male]
plt.pie([-sum(male), sum(female)], labels=['남자', '여자'], startangle=90, autopct='%.1f%%')
plt.legend()
```
